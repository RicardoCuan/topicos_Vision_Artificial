## Parte 0: Setup

```python
# Importar las librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Parte 1

Leer un imagen de su computadora se recomienda que esta imagen sea en formato horizontal o landscape. Convertir la imagen a formato de MatPlotLib (RGB).

```python
# Cargar una iagen
img_OpenCV = cv2.imread('img/pikachu.jpg')

# Dividir la imagen en sus 3 canales (b, g, r):
h,w,c = img_OpenCV.shape

# Tamaño de la imagen
dsize = (int(w*0.250), int(h*0.250))

# Re-dimensionar la imagen
img_OpenCV = cv2.resize(img_OpenCV, dsize)

b, g, r = cv2.split(img_OpenCV)
h, w, c = img_OpenCV.shape

# Combinar los canales pero en orden RGB
img_matplotlib = cv2.merge([r, g, b])

print("Dimensiones de la imagen - Alto: {}, Ancho: {}, Canales: {}".format(h, w, c))

# Combinar los canales pero en orden RGB
img_matplotlib = cv2.merge([r, g, b])



# RGB

# Obtener los canales, similar a cv2.split()
B = img_OpenCV[:, :, 0]
G = img_OpenCV[:, :, 1]
R = img_OpenCV[:, :, 2]

# Transformar la imagen BGR a RGB usando Numpy :
# ::-1 iterable que imprime los elementos en orden inverso
img_RGB = img_OpenCV[:, :, ::-1]




# Ver ambas imágenes(img_OpenCV y img_matplotlib) usando matplotlib
# Esta mostrará la imagen con los colores err´oneos
plt.subplot(121)
plt.imshow(img_OpenCV)
plt.title('img OpenCV')

# Esto mostrará los colores verdaderos
plt.subplot(122)
plt.imshow(img_matplotlib)
plt.title('img matplotlib')
plt.show()
```

## Parte 2

Crear una imagen utilizando NumPy y asignar al fondo el color (229, 93, 155), crear un recuadro en dicha imagen que debe ser del siguiente color (212, 245, 0) , esta imagen debe ser en formato horizontal o apaisada.

```python
# Creación de una imagen con dimensiones h x w y 3 canales de color
bgr = np.zeros((h, w, 3), dtype=np.uint8)

# Asignar un color a todos los píxeles de la imagen
bgr[:,:,:] = (229, 93, 155)

# Asignar un color a una sección de la imagen
bgr[50:75,100:200,:] = (212, 245, 0)


# Mostrar imagen en un plot
plt.subplot(121)
plt.imshow(bgr)
plt.title('img BGR')
plt.show()
```

## Parte 3

Concatene verticalmente las 3 imágenes (RGB, NumPy, BGR) utilizando la función np.concatenate() y adjunte la imagen generada a su informe.

```python
# Concatenar Imagenes Horizontalmente
# (img_OpenCV a la izquierda de img_matplotlib):
img_concats = np.concatenate((img_OpenCV, img_matplotlib, bgr), axis=0)

# Mostrar la imagen concatenada
cv2.imshow('bgr image and rgb image', img_concats)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
