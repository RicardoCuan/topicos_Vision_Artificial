## Parte 0

Instalación de librerías y cargar imágenes.

```python
# Importar las librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargar una iagen
image = cv2.imread('img/pikachu.jpg')

# Dividir la imagen en sus 3 canales (b, g, r):
h,w,c = image.shape

# Tamaño de la imagen
dsize = (int(w*0.250), int(h*0.250))

# Re-dimensionar la imagen
image = cv2.resize(image, dsize)
```

## Parte 1

Aplique para una imagen de su elección un filtro arbitrario de tamañ 6x6 y cuyo valor en cada elemento sea 1/37

```python
def show_with_matplotlib_6x6(color_img, title, pos):
    """Mostar una imagen usando MatPlotLib"""

    # Convertir imagen BGR a RGB
    img_RGB = color_img[:, :, ::-1]
    ax = plt.subplot(6, 6, pos)
    plt.imshow(img_RGB)
    plt.title(title)
    plt.axis('off')
```

```python
for i in range(1, 37):
    show_with_matplotlib_6x6(image, '', i)
```

## Parte 2

Aplique para la misma imagen un filtro arbitrario de tamaño 3x3 y cuyo valor en cada elemento sea 1/7 .

```python
def show_with_matplotlib_3x3(color_img, title, pos):
    """Mostar una imagen usando MatPlotLib"""

    # Convertir imagen BGR a RGB
    img_RGB = color_img[:, :, ::-1]
    ax = plt.subplot(3, 3, pos)
    plt.imshow(img_RGB)
    plt.title(title)
    plt.axis('off')
```

```python
for i in range(1, 10):
    show_with_matplotlib_3x3(image, '', i)
```

## Parte 3

Seleccione 2 de los filtros mostrados en el ejemplo y aplíquelos en su imagen 3 veces utilizando par´ametros diferentes para el filtro en cada aplicación

```python
# crear un objeto figura con las dimensiones adecuadas
plt.figure(figsize=(12, 6))
plt.suptitle("Técnicas de Smoothing", fontsize=14, fontweight='bold')

# Creamos el kernel para el suavizado de la imagen
# En este caso se crea un kernel (10,10)
kernel_averaging_10_10 = np.ones((10, 10), np.float32) / 100

# kernel_averaging_5_5 = np.ones((5, 5), np.float32)/25
kernel_averaging_5_5 = np.array([[0.04, 0.04, 0.04, 0.04, 0.04],
[0.04, 0.04, 0.04, 0.04, 0.04],
[0.04, 0.04, 0.04, 0.04, 0.04],
[0.04, 0.04, 0.04, 0.04, 0.04],
[0.04, 0.04, 0.04, 0.04, 0.04]])
print("kernel: {}".format(kernel_averaging_5_5))

# La funcioón cv2.filter2D() aplica el filtro linea arbitrario
smooth_image_f2D_5_5 = cv2.filter2D(image, -1, kernel_averaging_5_5)
smooth_image_f2D_10_10 = cv2.filter2D(image, -1, kernel_averaging_10_10)
```

```python
# La funcion cv2.blur() suaviza una imagen
smooth_image_b = cv2.blur(image, (10, 10))

# Filtro de Suavizado
smooth_image_bfi = cv2.boxFilter(image, -1, (10, 10), normalize=True)

# Filtro Gaussiano
smooth_image_gb = cv2.GaussianBlur(image, (9, 9), 0)

# Filtro de difuminación de imágenes
smooth_image_mb = cv2.medianBlur(image, 9)

# Reducción de ruido
smooth_image_bf = cv2.bilateralFilter(image, 5, 10, 10)
smooth_image_bf_2 = cv2.bilateralFilter(image, 9, 200, 200)
```

## Parte 4

Muestre en una sola ventana la imagen original m´as el resultado de aplicar los filtros, esta gr´afica debe contener 9 elementos.

```python
# Graficar las images:
show_with_matplotlib_3x3(image, "original", 1)
show_with_matplotlib_3x3(smooth_image_f2D_5_5, "filter2D 5x5", 2)
show_with_matplotlib_3x3(smooth_image_f2D_10_10, "filter2D 10x10", 3)
show_with_matplotlib_3x3(smooth_image_b, "blur", 4)
show_with_matplotlib_3x3(smooth_image_bfi, "boxFilter", 5)
show_with_matplotlib_3x3(smooth_image_gb, "GaussianBlur", 6)
show_with_matplotlib_3x3(smooth_image_mb, "medianBlur", 7)
show_with_matplotlib_3x3(smooth_image_bf, "bilateralFilter", 8)
show_with_matplotlib_3x3(smooth_image_bf_2, "bilateralFilter", 9)

# Mostrar los resultados de la apliación en lugar de un alto mando
plt.show()
```
