## Parte 0: Setup

Se importan las librerías a utilizar y se hace el setup requerido para las próximas partes

```python
# importar librerías
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline


## IMAGEN ##

# leer imagen
im_original = cv2.imread('img/pikachu.jpg')  

# Dimensiones de la Imagen
h,w,c = im_original.shape

# Tamaño de la imagen
dsize = (int(w*0.50), int(h*0.50))

# Re-dimensionar la imagen
im = cv2.resize(im_original, dsize)

# Nuevas alturas h y ancho w
h, w = (h // 2, w // 2)
```

## Parte 1: Parche 1 - Centro

Una modificación de color mediante la técnica del parche, el área modificar debe tener un forma cuadrada y 
ubicarse en el centro de la imagen.

```python
# Tamaño de la caja
box_size = 100

# Colocar parche de color 89, 112, 22 en el centro de la imagen 
im[h//2 - box_size//2 : h//2 + box_size//2, w//2 - box_size//2 : w // 2 + box_size//2] = (89, 112, 22)  
```

## Parte 2: Parche 2 - Inferior derecha

Una modificación de color mediante la técnica del parche de forma cuadrada ubicada en la esquina inferior derecha con el siguiente color (24, 185, 128) .

```python
# Tamaño de la caja
box_size = 100

# Colocar parche del color 24, 185, 128 en la esquina inferior derecha
im[h-box_size:h, w-box_size:w] = (24, 185, 128)
```

## Parte 3: Parche 3 - Superior derecha

Crear un parche de una zona cuadrada de la imagen y ubicarlo en la esquina superior derecha de la imagen.

```python
# Corte a la imagen original
im_cut = im_original[320:660, 300:1000]

# Redimensionar la imagen cortada
im_cut = cv2.resize(im_cut, (w//4, h//4))

# dimensiones de la imagen cortada
cut_h, cut_w, cut_c = im_cut.shape  

# Colocar el parche
im[0:cut_h,w-cut_w:w] = im_cut
```

## Parte 4: Mostrar imagen

Todas las modificaciones se deben ver en una sola imagen

```python
# Mostrar imagen en ventana
cv2.imshow("Imagen Modificada", im)

# Establecer espera
cv2.waitKey(0)

# Para destruir todas las ventanas creadas
cv2.destroyAllWindows()
```
